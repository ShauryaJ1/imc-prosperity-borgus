```python
import itertools
import numpy as np
```

```python
rates = np.array(
    [[1,1.45,0.52,0.72],
    [0.7,1,0.31,0.48],
    [1.95,3.1,1,1.49],
    [1.34,1.98,0.64,1]]
)
```

```python
prices = {0:"Snowballs",1:"Pizza",2:"Silicon Nuggets",3:"SeaShells"}
```

```python
def eval_seq(seq):
    if len(seq) == 0:
        return 1
    else:
        initial = 1
        for product,next_product in zip(seq,seq[1:]):
            initial *= rates[product][next_product]
        return initial            
```

```python
eval_seq([3,2,1,0,3])
```

```python
four_combos = list(itertools.product([0,1,2,3],repeat=4))
three_combos = list(itertools.product([0,1,2,3],repeat=3))
two_combos = list(itertools.product([0,1,2,3],repeat=2))
one_combos = list(itertools.product([0,1,2,3],repeat=1))
```

```python
all_combos = four_combos + three_combos + two_combos + one_combos
```

```python
all_combos = [list(combo) for combo in all_combos]
```

```python
all_combos
```

```python
len(all_combos)
```

```python
def max_seq(sequences):
    best_seq = None
    best_value = 0
    for seq in sequences:
        value = eval_seq([3] + seq + [3])
        if value > best_value:
            best_value = value
            best_seq = seq
    return best_seq, best_value
```

```python
max_seq(all_combos)
```
